# Training

## 1. Fungsi on the fly dataset

```python
import os
import glob
import random
import pandas as pd
import torch
from torch.utils.data import Dataset
import torchvision.transforms as transforms
from trdg.generators import GeneratorFromStrings

class OnTheFlyTRDGDataset(Dataset):
    def __init__(self, tsv_path, fonts_dir):
        # 1. Load data teks dari TSV (Sama seperti kodemu yang lama)
        self.df = pd.read_csv(
                    tsv_path, 
                    sep="\t", 
                    header=None, 
                    names=["id", "lang", "text"], 
                    quoting=3, 
                    on_bad_lines="skip"
                )
        self.df = self.df[self.df["id"] != "id"].reset_index(drop=True)
        
        # 2. Ambil semua path font
        self.all_fonts = glob.glob(os.path.join(fonts_dir, "*.ttf")) + glob.glob(os.path.join(fonts_dir, "*.otf"))
        if not self.all_fonts:
            raise FileNotFoundError(f"Tidak ada font ditemukan di folder {fonts_dir}")
            
        # 3. Transformasi untuk mengubah PIL Image hasil TRDG menjadi Tensor PyTorch
        # Serta normalisasi ukuran tinggi (height) gambar, misal diseragamkan ke 64 piksel
        self.transform = transforms.Compose([
            transforms.Grayscale(num_output_channels=1), # HTR biasanya lebih stabil di grayscale (1 channel)
            transforms.Resize((64, 256)),               # Menyamakan ukuran (Tinggi=64, Lebar=256) untuk batching
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))         # Normalisasi nilai piksel ke [-1, 1]
        ])

    def __len__(self):
        # Total baris kalimat yang tersedia (124k)
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        text_label = str(row["text"])
        
        # --- Copy-Paste Parameter Random Milikmu di Sini ---
        random_font = random.choice(self.all_fonts)
        skew_angle = random.randint(0, 8)
        dist_type = random.randint(0, 3)
        dist_ori = random.randint(0, 2)
        bg_type = random.randint(0, 1)
        img_size = random.randint(48, 96)
        
        # Inisialisasi generator TRDG langsung di dalam memori RAM
        gen = GeneratorFromStrings(
            [text_label],
            fonts=[random_font],
            skewing_angle=skew_angle,
            random_skew=True,
            distorsion_type=dist_type,
            distorsion_orientation=dist_ori,
            background_type=bg_type,
            size=img_size,
            fit=True,
        )
        
        # Ambil gambar PIL (Menggunakan fix utama darimu)
        img, _ = next(iter(gen))
        
        # Transformasikan gambar menjadi Tensor PyTorch
        img_tensor = self.transform(img)
        
        # Kembalikan pasangan data: (Tensor Gambar, Label Teks Asli)
        return img_tensor, text_label
```

## 2. Inisialisasi dataset

```python
from torch.utils.data import DataLoader

dataset_train = OnTheFlyTRDGDataset(tsv_path="corpus_1.tsv", fonts_dir="hw_fonts")

# 2. Buat DataLoader dengan multi-processing (num_workers)
train_loader = DataLoader(
    dataset_train, 
    batch_size=64,       # Sesuaikan dengan VRAM GPU laptopmu (mulai dari 32 atau 64)
    shuffle=True,        # Selalu acak urutan kalimat setiap epoch
    num_workers=4,       # PENTING: CPU akan men-generate gambar di background via multi-thread
    pin_memory=True      # Mempercepat transfer data dari RAM ke VRAM GPU
)
```

```python
import os
import re
import pandas as pd

# 1. Tentukan path file
file_path = "corpus_1.tsv"

if not os.path.exists(file_path):
    print(f"Error: File '{file_path}' tidak ditemukan. Pastikan posisinya satu folder dengan skrip ini.")
else:
    # 2. Load dataset (mengikuti konfigurasi yang kamu gunakan sebelumnya)
    print("Sedang membaca dataset...")
    df = pd.read_csv(
        file_path,
        sep="\t",
        header=None,
        names=["id", "lang", "text"],
        on_bad_lines="skip",
    )
    df = df[df["id"] != "id"].reset_index(drop=True)

    # Pastikan tidak ada nilai kosong (NaN) di kolom text agar regex tidak error
    df["text"] = df["text"].astype(str)

    # 3. Gabungkan semua teks untuk mencari angka unik yang eksis
    gabungan_teks = "".join(df["text"].tolist())
    angka_ditemukan = sorted(list(set(re.findall(r"\d", gabungan_teks))))

    # 4. Hitung jumlah kalimat yang mengandung angka
    mask_mengandung_angka = df["text"].str.contains(r"\d", regex=True)
    jumlah_kalimat_berangka = mask_mengandung_angka.sum()
    total_kalimat = len(df)
    persentase = (jumlah_kalimat_berangka / total_kalimat) * 100

    # 5. Cetak Hasil Analisis
    print("\n" + "=" * 40)
    print("          HASIL ANALISIS ANGKA          ")
    print("=" * 40)

    if angka_ditemukan:
        print(f"✓ Angka yang terdeteksi: {', '.join(angka_ditemukan)}")
        print(f"✓ Jumlah semua angka unik: {len(angka_ditemukan)} dari 10 digit (0-9)")
        print(
            f"✓ Total kalimat mengandung angka: {jumlah_kalimat_berangka} dari {total_kalimat} kalimat ({persentase:.2f}%)"
        )

        # Cek apakah ada digit 0-9 yang absen/hilang
        semua_digit = set(string.digits if "string" in locals() else "0123456789")
        digit_hilang = sorted(list(semua_digit - set(angka_ditemukan)))
        if digit_hilang:
            print(
                f"⚠ PERINGATAN: Ada digit yang TIDAK ditemukan di dataset: {', '.join(digit_hilang)}"
            )
        else:
            print(
                "✓ Aman! Semua digit dari 0 sampai 9 lengkap tersedia di dataset."
            )

        # Tampilkan 5 contoh sampel kalimatnya
        print("\nContoh kalimat yang mengandung angka di datasetmu:")
        sampel_kalimat = df[mask_mengandung_angka]["text"].head(5)
        for i, kalimat in enumerate(sampel_kalimat, 1):
            print(f"  {i}. {kalimat}")

    else:
        print("❌ Tidak ditemukan angka (0-9) sama sekali di kolom 'text' datasetmu.")

    print("=" * 40)
```

# Training

```python
import os
import string
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

# =====================================================================
# 1. MEMBUAT KAMUS KARAKTER (VOCABULARY) & FUNGSI ENCODER
# =====================================================================
# Kumpulkan semua karakter yang mungkin muncul di dataset-mu
# Huruf besar, huruf kecil, angka 0-9, dan tanda baca utama
all_chars = string.ascii_letters + string.digits + " .,()?!-"

# CTC Loss mewajibkan indeks 0 dipakai untuk karakter '[BLANK]'
char_map = {"[BLANK]": 0}
for idx, char in enumerate(all_chars):
    char_map[char] = idx + 1

NUM_CLASSES = len(char_map)


# Fungsi untuk mengubah teks string menjadi tensor angka
def encode_labels(labels_list, char_map):
    targets = []
    target_lengths = []
    for label in labels_list:
        # Ubah tiap huruf jadi angka berdasarkan kamus
        encoded = [char_map[c] for c in label if c in char_map]
        targets.extend(encoded)
        target_lengths.append(len(encoded))
    return torch.tensor(targets, dtype=torch.long), torch.tensor(
        target_lengths, dtype=torch.long
    )


# =====================================================================
# 2. ARSITEKTUR MODEL CRNN (RCNN UNTUK TEKS)
# =====================================================================
class CRNNModel(nn.Module):

    def __init__(self, num_classes, hidden_size=256):
        super(CRNNModel, self).__init__()

        # Fitur Visual (CNN) - Input gambar grayscale (1 channel)
        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # Ukuran mengecil setengah
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d((2, 1), stride=(2, 1)),  # Perkecil tinggi saja
            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d((2, 1), stride=(2, 1)),
        )

        # Mengubah tinggi gambar menjadi 1 agar bisa dibaca dari kiri ke kanan
        self.adaptive_pool = nn.AdaptiveAvgPool2d((1, None))

        # Fitur Urutan/Waktu (Bidirectional LSTM)
        self.rnn = nn.LSTM(
            512, hidden_size, bidirectional=True, num_layers=2, batch_first=True
        )

        # Layer Output (Prediksi Karakter)
        self.fc = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, x):
        # 1. Ekstrak Fitur Visual
        features = self.cnn(x)  # Shape: [Batch, 512, Height, Width]
        features = self.adaptive_pool(features)  # Shape: [Batch, 512, 1, Width]

        # 2. Reshape agar pas masuk ke LSTM
        features = features.squeeze(2)  # Buang dimensi tinggi -> [Batch, 512, Width]
        features = features.permute(
            0, 2, 1
        )  # Tukar posisi -> [Batch, Width, 512]

        # 3. Masuk ke LSTM untuk belajar pola urutan kata
        rnn_out, _ = self.rnn(features)  # Shape: [Batch, Width, Hidden*2]

        # 4. Prediksi distribusi karakter di setiap kolom gambar
        out = self.fc(rnn_out)  # Shape: [Batch, Width, Num_Classes]

        # CTC Loss di PyTorch minta format: [Width/Time_Steps, Batch, Num_Classes]
        out = out.permute(1, 0, 2)
        return out


# =====================================================================
# 3. SETUP PARAMETER & RUNNING PIPELINE
# =====================================================================
BATCH_SIZE = 16  # Tetap gunakan 16 dulu agar laptop Asus TUF tidak kepanasan
LR = 0.001
TOTAL_EPOCHS = 50
CHECKPOINT_PATH = "checkpoint_ocr_tuf.pt"

# Load Dataset Anda
dataset_train = OnTheFlyTRDGDataset(tsv_path="corpus_1.tsv", fonts_dir="hw_fonts")
train_loader = DataLoader(
    dataset_train,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=0,
    pin_memory=True,
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Training akan berjalan di: {device}")

# Inisialisasi Model CRNN Asli & CTC Loss
model = CRNNModel(num_classes=NUM_CLASSES).to(device)
criterion = nn.CTCLoss(blank=0, zero_infinity=True)
optimizer = optim.Adam(model.parameters(), lr=LR)

# Logika Load Checkpoint (Sistem Resume)
start_epoch = 0
global_step = 0

if os.path.exists(CHECKPOINT_PATH):
    print(
        f"➔ Menemukan checkpoint lama di '{CHECKPOINT_PATH}'. Mencoba me-restore progres..."
    )
    checkpoint = torch.load(CHECKPOINT_PATH)
    model.load_state_dict(checkpoint["model_state_dict"])
    optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
    start_epoch = checkpoint["epoch"] + 1
    global_step = checkpoint["global_step"]
    print(f"✓ Berhasil melanjutkan training dari Epoch ke-{start_epoch}!")
else:
    print(
        "➔ Tidak ada checkpoint ditemukan. Memulai training dari awal (Epoch 0)."
    )

# =====================================================================
# 4. TRAINING LOOP UTAMA (SUDAH AKTIF)
# =====================================================================
for epoch in range(start_epoch, TOTAL_EPOCHS):
    model.train()
    running_loss = 0.0

    print(f"\n=== Menggila di Epoch {epoch+1}/{TOTAL_EPOCHS} ===")

    for batch_idx, (images, labels) in enumerate(train_loader):
        images = images.to(device)

        # 1. Encode teks label menjadi angka tensor
        targets, target_lengths = encode_labels(labels, char_map)
        targets = targets.to(device)
        target_lengths = target_lengths.to(device)

        # 2. Reset Gradient Optimizer
        optimizer.zero_grad()

        # 3. Forward Pass ke Model CRNN
        outputs = model(images)  # Output shape: [Width, Batch, Num_Classes]

        # 4. Hitung panjang input (lebar prediksi) dari model untuk CTC Loss
        input_lengths = torch.full(
            size=(images.size(0),), fill_value=outputs.size(0), dtype=torch.long
        ).to(device)

        # 5. Hitung CTC Loss (Wajib di-log_softmax terlebih dahulu)
        loss = criterion(
            outputs.log_softmax(2), targets, input_lengths, target_lengths
        )

        # 6. Backward Pass & Update Bobot Model
        loss.backward()
        optimizer.step()

        global_step += 1
        running_loss += loss.item()

        # Cetak log setiap 50 batch untuk memantau pergerakan loss asli
        if batch_idx % 50 == 0:
            print(
                f"Batch {batch_idx}/{len(train_loader)} - Loss Saat Ini: {loss.item():.4f}"
            )

    # --- 5. SIMPAN CHECKPOINT SETIAP AKHIR EPOCH ---
    print(f"➔ Mengamankan progres Epoch {epoch+1}...")
    torch.save(
        {
            "epoch": epoch,
            "global_step": global_step,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
        },
        CHECKPOINT_PATH,
    )
    print(
        "✓ Checkpoint tersimpan! Aman untuk mematikan laptop jika diperlukan."
    )

print("\n Training Selesai 100%! Model siap digunakan untuk HTR.")
```

# Test model

```python
import os
import string
import torch
from PIL import Image
import torchvision.transforms as transforms
import torchvision.transforms.functional as F  # Untuk padding pintar
import cv2  # Tambahan untuk binarisasi gambar
import numpy as np  # Tambahan untuk konversi matriks

# 1. SETUP DEKODER (Mengubah angka kembali jadi teks)
all_chars = string.ascii_letters + string.digits + " .,()?!-"
char_map_inv = {i + 1: char for i, char in enumerate(all_chars)}
char_map_inv[0] = ""  # Blank character untuk CTC


def ctc_decode(output):
    """Fungsi Greedy Decoder untuk CTC"""
    arg_maxes = torch.argmax(output, dim=2)  # Shape: [Width, Batch]
    arg_maxes = arg_maxes.permute(1, 0)  # Tukar ke [Batch, Width]

    decoded_text = ""
    last_char = None

    for idx in arg_maxes[0]:
        char_idx = idx.item()
        if char_idx != 0 and char_idx != last_char:
            decoded_text += char_map_inv[char_idx]
        last_char = char_idx

    return decoded_text


# 2. LOAD MODEL & CHECKPOINT
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Pastikan CRNNModel dan NUM_CLASSES sudah didefinisikan di cell sebelumnya
model_test = CRNNModel(num_classes=NUM_CLASSES).to(device)

# Ganti sesuai nama file checkpoint yang kamu gunakan saat ini
CHECKPOINT_PATH = "checkpoint_ocr_tuf.pt"

if os.path.exists(CHECKPOINT_PATH):
    # Menggunakan weights_only=False agar tidak memicu warning kuning
    checkpoint = torch.load(
        CHECKPOINT_PATH, map_location=device, weights_only=False
    )
    model_test.load_state_dict(checkpoint["model_state_dict"])
    model_test.eval()  # Mode evaluasi (matikan dropout/batchnorm)
    print(f"✓ Model dimuat dari Epoch {checkpoint['epoch']+1}")
else:
    print(f"❌ Checkpoint '{CHECKPOINT_PATH}' tidak ditemukan!")


# 3. FUNGSI PREDIKSI GAMBAR (VERSI ANTI-GEPENG + BINARISASI)
def predict_image(image_path):
    # Langkah 1: Baca gambar menggunakan OpenCV dalam mode Grayscale
    img_cv = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Langkah 2: Jalankan Otsu's Thresholding
    # Ini akan membuat background kertas jadi putih bersih (255) dan tinta tulisan jadi hitam pekat (0)
    _, img_bin = cv2.threshold(
        img_cv, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )

    # Langkah 3: Kembalikan dari format OpenCV (Numpy) ke PIL Image agar sisa kode di bawah tidak patah
    img = Image.fromarray(img_bin)

    target_h = 64
    target_w = 256

    # Hitung lebar baru agar aspek rasio asli tulisan tangan tetap terjaga
    orig_w, orig_h = img.size
    w_percent = target_h / float(orig_h)
    new_w = int(float(orig_w) * float(w_percent))

    # Langkah A: Resize tinggi murni ke 64, lebar mengikuti secara proporsional
    img_resized = img.resize((new_w, target_h), Image.Resampling.LANCZOS)

    # Langkah B: Logika Cerdas agar pas masuk cetakan (64x256)
    if new_w < target_w:
        # Jika hasil resize kurang lebar dari 256, tambahkan sisa ruangnya dengan warna putih (255)
        padding_needed = target_w - new_w
        img_final = F.pad(img_resized, (0, 0, padding_needed, 0), fill=255)
    else:
        # Jika terlalu panjang ke samping, terpaksa di-resize lebarnya saja ke 256
        img_final = img_resized.resize(
            (target_w, target_h), Image.Resampling.LANCZOS
        )

    # Langkah C: Konversi ke Tensor dan Normalisasi
    transform = transforms.Compose(
        [transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))]
    )
    img_tensor = transform(img_final).unsqueeze(0).to(device)

    # Langkah D: Jalankan Forward Pass ke Model CRNN
    with torch.no_grad():
        output = model_test(img_tensor)
        probs = output.log_softmax(2)

    return ctc_decode(probs)


# --- CARA PAKAI ---
# Pastikan nama file gambar yang ingin kamu uji sudah sesuai di bawah ini
nama_file_tes = "test_3_rsz.png"

if os.path.exists(nama_file_tes):
    result = predict_image(nama_file_tes)
    print(f"Hasil Prediksi Model: {result}")
else:
    print(
        f"❌ File gambar '{nama_file_tes}' tidak ditemukan di folder kerja."
    )
```
