```python
import os
import glob

FOLDER_PATH = "D:/Github-Grinder/teks-dataset/manual_labeling/segmentation"

def rename_dataset_files(folder_path):
    files = sorted(glob.glob(os.path.join(folder_path, "*.jpg")))
    
    if not files:
        print("Tidak ada file .jpg yang ditemukan di folder tersebut.")
        return

    print(f"Menemukan {len(files)} file. Memulai proses rename...")
    for index, file_path in enumerate(files, start=1):
        ext = os.path.splitext(file_path)[1]
        
        new_name = f"{index:04d}{ext}"
        new_file_path = os.path.join(folder_path, new_name)
        
        os.rename(file_path, new_file_path)
        
    print("Semua file berhasil dirapikan!")

if __name__ == "__main__":
    rename_dataset_files(FOLDER_PATH)
```
