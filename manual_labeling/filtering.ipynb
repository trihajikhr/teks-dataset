```python
import os
import glob

FOLDER_PATH = "D:/Github-Grinder/teks-dataset/manual_labeling/segmentation"
START_NUM   = 291  # Ubah angka ini untuk menentukan nomor awal penamaan

def rename_dataset_files(folder_path, start_num):
    files = sorted(glob.glob(os.path.join(folder_path, "*.png")))
    
    if not files:
        print("Tidak ada file .jpg yang ditemukan di folder tersebut.")
        return

    print(f"Menemukan {len(files)} file.")
    print(f"Mengurutkan ulang dimulai dari nomor: {start_num:04d}")

    for index, file_path in enumerate(files, start=start_num):
        ext = os.path.splitext(file_path)[1]
        
        new_name = f"{index:04d}{ext}"
        new_file_path = os.path.join(folder_path, new_name)
        
        os.rename(file_path, new_file_path)
        
    print(f"Selesai! File terakhir di-rename menjadi: {index:04d}{ext}")

if __name__ == "__main__":
    rename_dataset_files(FOLDER_PATH, START_NUM)
```
