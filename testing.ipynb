```python
import os
import glob
import random
import pandas as pd
from trdg.generators import GeneratorFromStrings

output_dir = "corpus_dataset_1"
os.makedirs(output_dir, exist_ok=True)

df = pd.read_csv("corpus_1.tsv", sep="\t", header=None, names=["id", "lang", "text"], on_bad_lines="skip")
df = df[df["id"] != "id"].reset_index(drop=True)

all_fonts = glob.glob("hw_fonts/*.ttf") + glob.glob("hw_fonts/*.otf")

start = 0
end = 50

for _, row in df.iloc[start:end].iterrows():
    for i in range(50):
        gen = GeneratorFromStrings(
            [row["text"]],
            fonts=[random.choice(all_fonts)],
            skewing_angle=random.randint(0, 8),
            random_skew=True,
            distorsion_type=random.randint(0, 3),
            distorsion_orientation=random.randint(0, 2),
            background_type=random.randint(0, 1),
            size=random.randint(48, 96),
            fit=True,
        )
        img, _ = next(iter(gen))  # ← fix utama
        img.save(os.path.join(output_dir, f"{row['id']}_{i}.png"))

print(f"Done! Sesi {start}-{end} selesai.")
```
