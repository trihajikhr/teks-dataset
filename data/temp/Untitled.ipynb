Bersihkan data, pangkas yang terlalu panjang

```python
import pandas as pd

df = pd.read_csv("ind_sentences.tsv", sep="\t", header=None, names=["id", "lang", "text"], on_bad_lines="skip")

df["word_count"] = df["text"].str.split().str.len()
df_filtered = df[df["word_count"] <= 10].drop(columns=["word_count"])

df_filtered.to_csv("corpus.tsv", sep="\t", index=False)
```
