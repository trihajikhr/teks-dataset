```python
import cv2
from paddleocr import PaddleOCR

# 1. Inisialisasi PaddleOCR (rec=False artinya kita CUMA DETEKSI KOTAK, gak usah baca teksnya)
ocr = PaddleOCR(use_angle_cls=True, lang="id", rec=False)

# 2. Load gambar satu halaman penuh tulisan tangan lu
img_path = "foto_satu_halaman.png"
img = cv2.imread(img_path)

# Run deteksi kotak teks
result = ocr.ocr(img_path, cls=True)

# 3. Ekstrak koordinat kotak dan potong otomatis
count = 1
for line in result[0]:
    box = line[0]  # Mengambil 4 titik koordinat [[x1,y1], [x2,y2], [x3,y3], [x4,y4]]

    # Cari nilai min-max untuk bounding box
    x_min = int(min([pt[0] for pt in box]))
    x_max = int(max([pt[0] for pt in box]))
    y_min = int(min([pt[1] for pt in box]))
    y_max = int(max([pt[1] for pt in box]))

    # Potong gambar berdasarkan koordinat (Crop)
    cropped_line = img[y_min:y_max, x_min:x_max]

    # Simpan hasil potongan
    cv2.imwrite(f"dataset_asli/{count}.png", cropped_line)
    print(f"Berhasil memotong baris ke-{count}")
    count += 1
```
